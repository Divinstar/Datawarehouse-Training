```python
import pandas as pd

# Upload the 3 CSVs manually via file upload
from google.colab import files
uploaded = files.upload()
```

```python
# Load the data
stock_df = pd.read_csv("stock_summary.csv")
product_df = pd.read_csv("products.csv")
warehouse_df = pd.read_csv("warehouses.csv")

# Join stock + product
merged_df = pd.merge(stock_df, product_df, on="product_id", how="left")

# Join with warehouse
merged_df = pd.merge(merged_df, warehouse_df, on="warehouse_id", how="left")

# Add reorder flag
merged_df["needs_reorder"] = merged_df["total_stock"] < merged_df["reorder_level"]

final_inventory = merged_df[[
    "product_id", "product_name", "warehouse_id", "warehouse_name",
    "total_stock", "reorder_level", "needs_reorder"
]]


final_inventory.to_csv("final_inventory_view.csv", index=False)

files.download("final_inventory_view.csv")
```
